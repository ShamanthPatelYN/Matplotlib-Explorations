Load Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#Resize graph
plt.figure(figsize=(5,3) , dpi = 200)

# plt.plot(x , y , label='2x' , color = 'red' , linewidth = 2 , marker = '.' ,linestyle = '--' ,  markersize = 10 , markeredgecolor = 'black')

#Use shortcut notation
#fmt = '[color][maker][line]'
plt.plot(x,y, 'r*--' , label='2x')

#Line Number two

#Select Interval
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[0:5] , x2[0:5] ** 2 , color = 'black' , label = 'x^2')
plt.plot(x2[4:] , x2[4:]**2 , 'b--')

#add Title
plt.title('Our First Graph' , fontdict={'fontsize' : 15})

#X and Y Labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.xticks([0,1,2,3,4]) #indexing values
plt.yticks([0,2,4,6,8,10])

plt.legend()

#Save figure
plt.savefig('Mygraph.png' , dpi = 150)

plt.show()
```

Bar Graph

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(5,3), dpi=100)

bars = plt.bar(labels, values)

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
    
#bar[0].set_hatch('/')

plt.savefig('barchart.png', dpi=300)

plt.show()
```
